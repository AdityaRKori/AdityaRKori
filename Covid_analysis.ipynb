```python
df['Date Of Hospitalisation'] = pd.to_datetime(df['Date Of Hospitalisation'])
df['Date of Discharge/Death'] = pd.to_datetime(df['Date of Discharge/Death'])
df['length of stay'] = (df['Date of Discharge/Death'] - df['Date Of Hospitalisation']).dt.days
display(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the scatter plot
sns.regplot(x='Age', y='length of stay', data=df)

# Customize the plot (optional)
plt.title('Correlation between Age and Length of Stay')
plt.xlabel('Age')
plt.ylabel('Length of Stay (days)')

# Display the plot
plt.show()
```
