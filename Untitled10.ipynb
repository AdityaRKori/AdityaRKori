```python
# prompt: use the 'Blood Pressure' column and make two separate columns namely 'Blood Pressure Systolic' & 'Blood Pressure Diastolic'

import pandas as pd
df = pd.read_csv('/content/heart_attack_prediction_dataset_edit.csv')

# Split 'Blood Pressure' column into 'Blood Pressure Systolic' and 'Blood Pressure Diastolic'
df[['Blood Pressure Systolic', 'Blood Pressure Diastolic']] = df['Blood Pressure'].str.split('/', expand=True)

# Print the updated DataFrame
print(df.head())
```

```python
import pandas as pd

try:
    df[['Systolic_Blood Pressure', 'Diastolic_Blood Pressure']] = df['Blood Pressure'].str.split('/', expand=True)
    df['Systolic_Blood Pressure'] = pd.to_numeric(df['Systolic_Blood Pressure'], errors='coerce')
    df['Diastolic_Blood Pressure'] = pd.to_numeric(df['Diastolic_Blood Pressure'], errors='coerce')
    # To see the output, run the code.
    print(df.head())
except:
    print("Error: 'Blood Pressure' column not found or data format is incorrect.")
```

```python
# prompt: drop blood pressure column

try:
    df = df.drop('Blood Pressure', axis=1)
    print(df.head())
except KeyError:
    print("Error: 'Blood Pressure' column not found.")
```

```python
for i in ["Age","BMI","Blood Pressure Systolic","Blood Pressure Diastolic"]: # Removing columns with string values
    # Convert the column to numeric, handling errors by coercing to NaN
    df[i] = pd.to_numeric(df[i], errors='coerce')
    mean_i = df[i].mean()
    df[i].fillna(mean_i, inplace=True)
    print(df.shape)
    print(df.shape)
```

```python
import numpy as np
for i in df.select_dtypes(include = np.number).columns[:-1]:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[i] >= lower_bound) & (df[i] <= upper_bound)]
    print(df.shape)
```

```python
df['Diet'].fillna(df['Diet'].mode()[0], inplace = True)
df['Diet'] = df['Diet'].replace({'Unhelthy': 'Unhealthy','Healty': 'Healthy'})
```

```python
df['Sex'].fillna(df['Sex'].mode()[0], inplace = True)
df['Sex'] = df['Sex'].replace({'m': 'Male','f': 'Female'})
```

```python
encoded_data = pd.get_dummies(df, columns=['Patient ID', 'Sex', 'Diet', 'Country', 'Continent', 'Hemisphere'], # Removed the trailing space from 'Smoking Status '
                              prefix=['Patient ID', 'Sex', 'Diet', 'Country', 'Continent', 'Hemisphere']) # Removed the trailing space from 'Smoking Status '
df = encoded_data
print(df.head())
```
