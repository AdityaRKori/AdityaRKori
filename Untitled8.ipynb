```python
# prompt: replace unique values with mean values

import pandas as pd
import numpy as np
# Assuming the uploaded file is named 'healthcare_regression_dataset.csv'
try:
    df = pd.read_csv('/content/healthcare_regression_dataset.csv')

    # Identify numerical columns
    numerical_cols = df.select_dtypes(include=np.number).columns

    # Remove outliers using IQR method for each numerical column
    for col in numerical_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

    # Replace unique values in non-numerical columns with the mean of the column
    for col in df.columns:
        if not pd.api.types.is_numeric_dtype(df[col]):
            # Calculate the mean of numerical columns for replacement
            mean_value = df.select_dtypes(include=np.number).mean(numeric_only=True).mean() #Mean of means
            #Replace unique values with mean
            df[col] = df[col].apply(lambda x: mean_value if x in df[col].unique() else x)

    print(df)

except FileNotFoundError:
    print("Error: 'healthcare_regression_dataset.csv' not found. Please upload the file.")
except pd.errors.EmptyDataError:
    print("Error: 'healthcare_regression_dataset.csv' is empty.")
except pd.errors.ParserError:
    print("Error: Unable to parse 'healthcare_regression_dataset.csv'. Check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# prompt: get dummy values for each columns

import pandas as pd
import numpy as np
# Assuming the uploaded file is named 'healthcare_regression_dataset.csv'
try:
    df = pd.read_csv('/content/healthcare_regression_dataset.csv')

    # Identify numerical columns
    numerical_cols = df.select_dtypes(include=np.number).columns

    # Remove outliers using IQR method for each numerical column
    for col in numerical_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

    # Replace unique values in non-numerical columns with dummy values
    for col in df.columns:
        if not pd.api.types.is_numeric_dtype(df[col]):
            # Generate dummy values (e.g., using a sequence or random values)
            dummy_values = [f"Dummy_{i}" for i in range(len(df[col].unique()))]

            # Create a mapping for unique values
            value_mapping = dict(zip(df[col].unique(), dummy_values))

            # Replace values in the column
            df[col] = df[col].map(value_mapping)

    print(df)

except FileNotFoundError:
    print("Error: 'healthcare_regression_dataset.csv' not found. Please upload the file.")
except pd.errors.EmptyDataError:
    print("Error: 'healthcare_regression_dataset.csv' is empty.")
except pd.errors.ParserError:
    print("Error: Unable to parse 'healthcare_regression_dataset.csv'. Check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
