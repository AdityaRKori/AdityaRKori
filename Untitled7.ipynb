```python
# prompt: read the uploaded file

import pandas as pd

# Assuming the uploaded file is named 'data.csv'
try:
  df = pd.read_csv('/content/Health_insurance.csv')
  print(df)
except FileNotFoundError:
  print("Error: 'data.csv' not found. Please upload the file.")
except pd.errors.EmptyDataError:
  print("Error: 'data.csv' is empty.")
except pd.errors.ParserError:
  print("Error: Unable to parse 'data.csv'. Check the file format.")
```

```python
df = df.drop_duplicates(subset=['pid'])
```

```python
def remove_outliers_iqr(df,column):
  Q1=df[column].quantile(0.25)
  Q3=df[column].quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR
  df=df[(df[column]>lower_bound) & (df[column]<upper_bound)]
  return df
  df=remove_outliers_iqr(df,'age')
  df=remove_outliers_iqr(df,'bmi')
  df=remove_outliers_iqr(df,'children')
  df=remove_outliers_iqr(df,'charges')
  df()
```

```python
import numpy as np
```

```python
numeric_cols=df.select_dtypes(include=np.number).columns
for col in numeric_cols:
  df=remove_outliers_iqr(df,col)
```

```python
df.isnull().sum()
```

```python
columns_to_fill=['bmi','heart rate','glucose']
for col in columns_to_fill:
  df[col]=df[col].fillna(df[col].mean())
```

```python
category_cols=df.select_dtypes(include='object').columns
for col in category_cols:
  print(df[col].unique())
```

```python
df['sex'].fillna(df['sex'].mode()[0], inplace = True)
df['sex'] = df['sex'].replace({'mal': 'male','female': 'Female'})
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['sex', 'smoker', 'region'])
```

```python
df.head()
```

```python
df.head()
```

```python
#normalize numerial columns
min_max_values = {}

cols = ['age', 'bmi', 'heart rate', 'Creatinine', 'glucose']

for i in cols:
    min_val = df[i].min()
    max_val = df[i].max()
    df[i] = (df[i] - min_val) / (max_val - min_val)
    min_max_values[i] = (min_val, max_val)
```

```python
import json
with open('min_max_values.json', 'w') as json_file:
    json.dump(min_max_values, json_file)
```

```python
x=df.drop(columns=['charges','pid','sex_male','smoker_yes'],axis=1)
y=df['charges']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
y_pred = reg.predict(x_test)
```

```python
#compute the rense, r2 score
from sklearn.metrics import mean_squared_error,r2_score
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print("RMSE:", rmse)
print("R2 Score:", r2)
```
